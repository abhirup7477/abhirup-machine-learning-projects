## **Task-6: Use NumPy to calculate basic statistic**

### price

```python
price_mean = np.mean(df["price"])
price_median = np.median(df["price"])
price_min = np.min(df["price"])
price_max = np.max(df["price"])
price_std = np.std(df["price"])

print("Price statistics:")
print("mean =",price_mean)
print("median =",price_median)
print("min value =",price_min)
print("max value =",price_max)
print("standard deviation value =",price_std)
```

### freight_value

```python
freight_mean = np.mean(df["freight_value"])
freight_median = np.median(df["freight_value"])
freight_min = np.min(df["freight_value"])
freight_max = np.max(df["freight_value"])
freight_std = np.std(df["freight_value"])

print("Freight value statistics:")
print("mean =",freight_mean)
print("median =",freight_median)
print("min value =",freight_min)
print("max value =",freight_max)
print("standard deviation value =",freight_std)
```

**Observation:**

The analysis reveals a significant right-skew in both variables, as the means are notably higher than the medians. Product prices show extreme variability, ranging from 0.85 to 6,735.00, with a high standard deviation indicating a diverse product catalog. Freight values are more consistent but include outliers up to 409.68, which likely represent bulky items or long-distance deliveries.

## **Task 16: Calculate the correlation between price, freight_value, and total_price. Visualise correlation using a heatmap.**

```python
import seaborn as sns

sns.heatmap(corr_matrix, annot=True)

plt.title("Correlation Heatmap using seaborn")
plt.show()
```

### Visualise correlation heatmap using matplotlib library

```python
plt.figure(figsize=(5, 5))
plt.imshow(corr_matrix, cmap="coolwarm", interpolation="none")
plt.colorbar()

plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        plt.text(i, j, round(corr_matrix.iloc[i, j], 2), ha="center", va="center", color="black")

plt.title("Correlation Heatmap")
plt.show()
```

* The correlation between price and freight_value is relatively low (0.42), indicating that expensive items do not always incur high shipping costs.

## **Task 22: Analyse if high shipping costs (freight_value) correlate with delayed deliveries**

### Overall correlation freight_value and delivery_time

```python
overall_corr = df["freight_value"].corr(df["delivery_time"])

print("Overall correlation freight_value and delivery_time:", overall_corr)
```

### Correlation for high shipping cost subset

```python
# Calculate high shipping cost threshold
high_shipping_threshold = np.percentile(df["freight_value"], 75)
print("high_shipping_threshold:", high_shipping_threshold)

# Filter out high shipping costs
high_shipping_records = df[df["freight_value"] > high_shipping_threshold]

# Calculate correlation
high_shipping_corr = high_shipping_records["freight_value"].corr(high_shipping_records["delivery_time"])

print("high_shipping_corr:", high_shipping_corr)
```

**Observation:**

* There is no strong link between shipping cost and delivery time.

* The overall correlation is quite low (0.21).

* After filtering out expensive shipping (75th percentile) and correlation with delivery time further dropped.

* This suggest paying high shipping cost does not mean getting faster delivery.
